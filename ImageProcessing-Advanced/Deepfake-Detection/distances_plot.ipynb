```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('distances.csv')
df.head()
```

```python
distances_real = df[df['label'] == 0]['distance'].to_numpy()
distances_real.shape
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12, 8
sns.distplot(distances_fake, bins=100, kde=True);
```
